Lectra de bases

```python
# Base de beneficiarios
names_benef = ['ID','Anio','CVE_ENT','NOM_ENT','CVE_MUN','NOM_MUN','CVE_LOC','NOM_LOC','MONTO_APOYO_TOTAL']
base_benef = pd.read_excel('Base_benef_completa.xlsx')
base_benef= base_benef.rename({'AÑO':'Anio'}, axis=1)
base_benef = base_benef.drop(['Estado',	'Municipio',	'Localidad'], axis=1)
base_tamanio_productor = pd.read_excel('Base_tamaño_productor.xlsx', sheet_name='BASE')
base_tamanio_productor.columns = ['ID', 'Tamanio_productor']
base_grado_marginacion = pd.read_excel('IM_2020.xlsx')
base_grado_marginacion = base_grado_marginacion[['ENT', 'MUN', 'LOC', 'GM_2020']]
```

```python
base_benef.head()
```

```python
base_tamanio_productor.head()
```

```python
base_grado_marginacion.head()
```

# Unión de bases

```python
df_merged = pd.merge(base_benef, base_tamanio_productor,  
                      left_on=['ID'], 
                      right_on=['ID'], how='left')

df_merged.head()
```

```python
df_merged = pd.merge(df_merged, base_grado_marginacion,
                     left_on=['CVE_ENT',	'CVE_MUN', 'CVE_LOC'], 
                     right_on=['ENT', 'MUN', 'LOC'], how='left')

df_merged = df_merged.drop(['ENT', 'MUN', 'LOC'], axis=1)
df_merged.head()
```

**Pegamos el producto**

```python
base_benef_producto = pd.read_excel('CONSOLIDADO.xlsx', sheet_name='BASE')
base_benef_producto.head()
```

```python
df_merged = pd.merge(df_merged, base_benef_producto,
                     left_on=['ID'], 
                     right_on=['ID'], how='left')
df_merged.head()
```

```python
df_merged.to_csv('base_beneficiarios_dashboard.csv', index=False, encoding="utf-8-sig")
```
