```python
# dataframe con la minima distancia
def closeCenter(x, y, xlabs, ylabs):
  '''
  Inputs:
    x,y   : data
    xlabs : tupla (lat, lon)
    ylabs : tupla (lat, lon)
  Outputs:
    dataframe : dataframe with closest centers
  '''  # rename datasets
  base1 = x
  centros = y
  # labels
  latx, lonx = xlabs[0], xlabs[1]
  laty, lony = ylabs[0], ylabs[1]
  data = []
  # minima distancia
  minDist = []
  i = 0
  for lat1, lon1 in zip(base1[latx], base1[lonx]):
    coord1 = (lat1, lon1)
    #print('coord1', coord1)
    dist = []
    for lat2, lon2 in zip(centros[laty], centros[lony]):
      coord2 = (lat2, lon2)
      #print('coord2', coord2)
      # calcula distancia
      d = geodesic(coord1, coord2).km
      # agrga a lista
      dist.append(d)
      # print
      print('Row: ', i)
      i += 1
    # min distance
    mindist = min(dist)
    # position min
    idx = pd.Series(dist).idxmin()
    # append data
    data.append(centros.iloc[idx,:])
    minDist.append(mindist)

  result = pd.concat(data, axis= 1).T

  return result
```

```python
# Filtros
anio = 2021
# base filtrada
base1_filt = base1[base1['Anio']==anio]
#base1_filt = base1_filt[base1_filt['Producto']==producto]

df_result = closeCenter(base1_filt, centros,
                    xlabs=('LAT_DECIMAL', 'LON_DECIMAL'),
                    ylabs=('LAT_DECIMAL', 'LON_DECIMAL'))
df_result.head()
```

```python
result = pd.concat([base1_filt.reset_index(),df_result.reset_index(drop=True)[['NUM','NOMBRE_UNIDAD','NOM_ENT','NOM_MUN','NOM_LOC']]], axis=1)
result.to_excel('resultado_2021.xlsx')
```
